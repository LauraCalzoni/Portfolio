# Prefiltering the image

### Optional part (4)

Here we have filtered the stereo images by different low-pass filters, before
carrying out the matching process. In particular, we have tried boh the Bilateral and the Non-Local Means filter, two non linear, edge-preserving smoothing filters. 

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def open_img(img):
    image=cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    copied_image= np.copy(image)
    plt.imshow(copied_image, cmap='gray', vmin=0, vmax=255)
    plt.show()
    
    return copied_image


#functions for SAD implementation
def AD_left(imgL, imgR, i, j, disp, m, n):
    S=np.abs(int(imgL[i+m, j+n])-int(imgR[i+m, j+n+disp]))
    
    return S


#function to calculate the mapping error
def Error( target, img, w, h, m, scale_factor, ig_border, d_max):
    error=0
    window=(w-2*m-ig_border-2*d_max)*(h-2*m-ig_border)
    for j in range(m+d_max, w-m-d_max-ig_border):
        for i in range(m, h-m-ig_border):
            if(abs(target[i, j]/scale_factor - img[i, j]/scale_factor)>1):
                error=error+1
    
    return (error/window)
```

###### Bilateral filter

```python
print("bilateral filtered images")
sigmaColor=5
sigmaSpace=5
kernel_size=int(np.ceil((3*sigmaSpace))*2 + 1)
BilateralL=cv2.bilateralFilter(imageL, kernel_size, sigmaColor, sigmaSpace)
BilateralR=cv2.bilateralFilter(imageR, kernel_size, sigmaColor, sigmaSpace)

# Plot the result
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(BilateralL ,cmap='gray',vmin=0,vmax=255)
plt.subplot(1,2,2)
plt.imshow(BilateralR ,cmap='gray',vmin=0,vmax=255)
plt.show()
```

###### Non-Local Means filter

Parameters:
- h = parameter which regulates the filter strenght. A high value better removes noise but alsoimage details
- N (template window size) = size in pixels of the template path that is used to compute weights (should be odd)
- S (searching window size) = size in pixels of the window that is used to compute the weighted average for the given pixel (should be odd)

```python
print("non local means filter")
h=10
N=7   
S=21  
NLM_L=cv2.fastNlMeansDenoising(imageL, h, N, S)
NLM_R=cv2.fastNlMeansDenoising(imageR, h, N, S)

# Plot the result
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(NLM_L ,cmap='gray',vmin=0,vmax=255)
plt.subplot(1,2,2)
plt.imshow(NLM_R ,cmap='gray',vmin=0,vmax=255)
plt.show()
```

Now the supporting window'size is chosen. As we have explained in the report, we have tried multiple values, seeking for the better one that fits each image. According to our experimental results K=3 best fits Map and Sawtooth, while K=4 should be selected for Tsukuba and Venus.

```python
#window shape
K=3

MapL=np.zeros((img_raw, img_column), dtype=np.uint8)
```

The filter's type must be selected and applied, by commenting out the other one.


```python
#imageL=BilateralL
#imageR=BilateralR

imageL=NLM_L
imageR=NLM_R
```

The stereo algorithm is the same of the baseline case, as well as the error function:

```python
#in order to get every pixels' position in the image we compute two loops 
disp_L=0

for i in range(K, img_raw-K-ignore_border):
    for j in range(K+disp_max, img_column-K-ignore_border-disp_max): 
        #We select the pixel and move along its row
        min_SAD_L=500000
        #moving along the epipolar line
        for d in range(disp_min, disp_max+1):
            SAD_L=0
            for m in range(-K, K+1): 
                for n in range(-K, K+1):
                    SAD_L =SAD_L+AD_left(imageL, imageR, i, j, d, m, n)
                                
            if(SAD_L<=min_SAD_L):  
                min_SAD_L=SAD_L
                disp_L=d
            
                    
        MapL[i,j]=np.asarray([disp_L*disp_scale])
       

print("Left map")
plt.imshow(MapL, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
mapping_error=Error(mapG,MapL,img_column,img_raw,K, disp_scale, ignore_border, disp_max)
#mapping_error=Error(sawtoothG,MapL,img_column,img_raw,K, disp_scale, ignore_border, disp_max)
#mapping_error=Error(tsukubaG,MapL,img_column,img_raw,K, disp_scale, ignore_border, disp_max)
#mapping_error=Error(venusG,MapL,img_column,img_raw,K, disp_scale, ignore_border, disp_max)


print("the mapping error with prefiltering is:\n", mapping_error)
```
