# Baseline Stereo Matching algorithm 

### Mandatory part 

The following piece of code represent the implementation of a baseline local stereo matching algorithm. 
For each point in the reference image, the window centered at that point is compared to all those centered at the points in the other image laying on the same row and within the provided disparity range. The comparison between windows is carried out according to the SAD dissimilarity function, as required. The window providing the lowest SAD determines the correspondence and thus the disparity value for the given point. 
The left image is taken as reference. 

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def open_img(img):
    image=cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    copied_image= np.copy(image)
    plt.imshow(copied_image, cmap='gray', vmin=0, vmax=255)
    plt.show()
    
    return copied_image


#functions for SAD implementation
def AD_left(imgL, imgR, i, j, disp, m, n):
    S=np.abs(int(imgL[i+m, j+n])-int(imgR[i+m, j+n+disp]))
    
    return S


#function to calculate the mapping error
def Error( target, img, w, h, m, scale_factor, ig_border, d_max):
    error=0
    window=(w-2*m-ig_border-2*d_max)*(h-2*m-ig_border)
    for j in range(m+d_max, w-m-d_max-ig_border):
        for i in range(m, h-m-ig_border):
            if(abs(target[i, j]/scale_factor- img[i, j]/scale_factor)>1):
                error=error+1
    
    return (error/window)
```

Now the supporting window'size is chosen. As we have explained in the report, we have tried multiple values, seeking for the better one that fits each image. According to our experimental results K=3 best fits Map and Sawtooth, while K=4 should be selected for Tsukuba and Venus.

```python
#window shape (multiple different values tried)
K=3  #to set according to the image loaded

MapL=np.zeros((img_raw, img_column), dtype=np.uint8)
```

```python
#in order to get every pixels' position in the image we compute two loops 
#the indexes'values start from K 
#the indexes are increased until the maximum useful value minus K otherwise it overflows
disp_L=0
disp=0
for i in range(K, img_raw-K-ignore_border):
    for j in range(K+disp_max, img_column-K-ignore_border-disp_max): 
        #We select the pixel and move along its row
        min_SAD=500000
        #moving along the epipolar line
        for d in range(disp_min, disp_max+1):
            SAD=0
            for m in range(-K, K+1): #m and n should span this range to be able to catch all the pixels starting from K
                for n in range(-K, K+1):
                    #compute the SAD
                    SAD=SAD+AD_left(imageL, imageR, i, j, d, m, n)
            
            #winner takes-it-all approach
            if(SAD<=min_SAD):
                min_SAD=SAD
                disp=d
        
        MapL[i,j]=np.asarray([disp*disp_scale])

print("Left map")
plt.imshow(MapL, cmap='gray', vmin=0, vmax=255)
plt.show()
```

In the end, the mapping error is computed, according to the Error function specified in the paper. The line of interest should be uncommented.
