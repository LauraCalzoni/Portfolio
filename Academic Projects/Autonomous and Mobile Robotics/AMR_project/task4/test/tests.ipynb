```python
import numpy as np
import math 

grid_size = (30, 15)
grid = np.zeros(grid_size)
agent_pos = (15,7)
grid[agent_pos]=8

# Generate n random indices
indices = np.random.randint(0, grid_size[0]*grid_size[1], size=10)
rows = indices // grid_size[1]
cols = indices % grid_size[1]

# Set the randomly selected elements to a certain number
value_to_set = -3  # Example value to set
for r, c in zip(rows, cols):
    if r!=6 and c!=3:
        grid[r, c] = value_to_set


def pixels_in_shadow():
       
        obstacle_map = np.where(grid == -3, 1, 0)

        obstacles = []
        for i in range(len(obstacle_map)):
            for j in range(len(obstacle_map[i])):
                if obstacle_map[i][j] == 1:
                    obstacles.append((i, j))
        
        # know we have a list with the obstacles positions, now for
        # each position i have tu calculate the ray of the shadow

        #print(obstacles)
        dark = []
        k=0.5 # coefficient to make the ray centered in the pixel
        for obstacle in obstacles: 

            if agent_pos[1]!=obstacle[1]:
                m = ((obstacle[0]+k) - (agent_pos[0]+k))/((obstacle[1]+k) - (agent_pos[1]+k))
                q = (agent_pos[0]+k)-(agent_pos[1]+k)*m

            if agent_pos[1]<obstacle[1]:

                for x in range(obstacle[1]+1,grid_size[1]):
                    xp = x+k
                    y = math.floor(m*xp+q)
                    if y<grid_size[0] and y>=0:                      
                        dark.append((y,x))
                
            if agent_pos[1]>obstacle[1]:

                for x in range(obstacle[1]-1,-1,-1):
                    xp = x+k
                    y = math.floor(m*xp+q)
                    if y<grid_size[0] and y>=0:                     
                        dark.append((y,x))

            if agent_pos[0]>obstacle[0] and agent_pos[1]==obstacle[1]:

                for y in range(obstacle[0]-1,-1,-1):
                    if y<grid_size[0] and y>=0:                  
                        dark.append((y,obstacle[1]))

            if agent_pos[0]<obstacle[0] and agent_pos[1]==obstacle[1]:

                for y in range(obstacle[0]+1,grid_size[0]):
                    dark.append((y,obstacle[1]))
        
        return dark

coordinate = pixels_in_shadow()

#print(grid)
print(coordinate)
for x, y in coordinate:
    grid[x][y] = 1

print(grid)
```
